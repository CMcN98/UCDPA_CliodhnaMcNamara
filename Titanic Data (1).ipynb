```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
%matplotlib inline
sns.set(color_codes=True)
```

```python
# Import Dataset
path_to_file = r"C:\Users\cmcnamara\Documents\DA\TitanicDataset.csv"
Titanic = pd.read_csv(path_to_file)
```

```python
# Dataframe
Titanic = pd.DataFrame(Titanic)
```

```python
# Dropping columns that are irrelevant
Titanic = Titanic.drop(['Cabin', 'Ticket', 'PassengerId'], axis = 1)
Titanic.head(5)
```

```python
# Renaming columns
Titanic = Titanic.rename(columns={"Sex" : "Gender", "Pclass" : "Ticket_Class", "Fare" : "Ticket_Price" })
Titanic.head(5)
```

```python
# Merging columns Sibsp and Parch to make column Family Size
Titanic["Family_Size"] = Titanic["SibSp"] + Titanic["Parch"]
```

```python
Titanic.head(5)
```

```python
# Dropping columns SibSp and Parch
Titanic = Titanic.drop(['SibSp', 'Parch'], axis = 1)
Titanic.head(5)
```

```python
# Splitting Name into two columns: First name and Last name
Titanic['fullNameSplitLength'] = Titanic['Name'].str.split().str.len()
Titanic['First_Name'] = Titanic['Last_Name'] = ''
Titanic.loc[
  Titanic['fullNameSplitLength'] >= 1,
  'First_Name'
] = Titanic.loc[Titanic['fullNameSplitLength'] >= 1]['Name'].str.split().str[0]
Titanic.loc[
  Titanic['fullNameSplitLength'] >= 2,
  'Last_Name'
] = Titanic.loc[Titanic['fullNameSplitLength'] >= 2]['Name'].str.split().str[-1]
```

```python
# Dropping column fullNameSplitLength
Titanic = Titanic.drop(['fullNameSplitLength'], axis = 1)
Titanic.head(5)
```

```python
# Checking for duplicate rows
Titanic.duplicated().sum()
```

```python
# Checking for missing or null values
print(Titanic.isnull().sum())
```

```python
# Checking for outliers of Age
sns.boxplot(x=Titanic['Age'])
```

```python
Titanic.describe()['Age']
```

```python
ax = Titanic.hist(column='Age', bins=25, grid=True, figsize=(12,8), color='#b886bf', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

    # Remove title
    x.set_title("Age range")

    # Set x-axis label
    x.set_xlabel("Age", labelpad=20, weight='bold', size=12)

    # Set y-axis label
    x.set_ylabel("", labelpad=20, weight='bold', size=12)
```

```python
# Creating a new column and sorting ages into groups
Titanic['Age_Range'] = np.where(Titanic['Age'] < 20, '0 - 19', 
         (np.where(Titanic['Age'] < 50, '20 - 49', '50 - 80')))
```

```python
Titanic.head(5)
```

```python
# Plot to show survival based on Age Range
# In survived axis: 0 = No and 1 = Yes
sns.countplot(x='Survived', hue='Age_Range', data=Titanic).set(title="Passengers survived based on Age Range", ylabel="Number");
```

```python
# Survival percentage based on gender
Titanic[['Gender', 'Survived']].groupby(['Gender'], as_index = False).mean()
```

```python
# Percentage of female that survived
female = Titanic.loc[Titanic.Gender == 'female']["Survived"]
Percentage_female = sum(female)/len(female)
print(Percentage_female)
```

```python
# Percentage of male that survived
male = Titanic.loc[Titanic.Gender == 'male']["Survived"]
Percentage_male = sum(male)/len(male)
print(Percentage_male)
```
